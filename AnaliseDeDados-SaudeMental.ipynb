```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
```

```python
data = pd.read_csv("data.csv")
```

```python
for coluna in enumerate(data.columns): print (coluna)
```

```python
data.rename(columns={'Age': 'IDADE',
                    'Gender': 'GENERO',
                    'family_history': 'HISTORICO_FAMILIAR',
                    'treatment': 'TEM_TRATAMENTO',
                    'work_interfere': 'INTERFERE_NO_TRABALHO',
                    'supervisor': 'SUPERVISAO_MEDICA'}, inplace = True)    
                     
```

```python
data.sample(40)
```

```python
plt.title('Gráfico de Idades de Pessoas com Doenças Mentais')
plt.ylabel('Quantidade de Pessoas')

data.query('IDADE >= 0 and IDADE <= 120').groupby('IDADE').size().plot(kind='bar', figsize=(18,10),color='#A020F0',rot=40)
```

```python
data.loc[(data.HISTORICO_FAMILIAR == 'Yes'), 'HISTORICO_FAMILIAR'] = 'Tem histórico'
data.loc[(data.HISTORICO_FAMILIAR == 'No'), 'HISTORICO_FAMILIAR'] = 'Não tem histórico'
```

```python
plt.title('Pessoas que tem um histórico familiar com doenças mentais')

data.groupby('HISTORICO_FAMILIAR').size().plot(kind='bar', figsize=(5,5),  color='#A020F0',rot=40)
```

```python
data.loc[(data.TEM_TRATAMENTO == 'Yes'), 'TEM_TRATAMENTO'] = 'Tem tratamento'
data.loc[(data.TEM_TRATAMENTO == 'No'), 'TEM_TRATAMENTO'] = 'Não tem tratamento'
```

```python
plt.title('Pessoas com doenças mentais que tem tratamento')

data.groupby('TEM_TRATAMENTO').size().plot(kind='bar', figsize=(5,5),  color='#A020F0',rot=40)
```

```python
data.loc[(data.INTERFERE_NO_TRABALHO == 'Often'), 'INTERFERE_NO_TRABALHO'] = 'Frequentemente'
data.loc[(data.INTERFERE_NO_TRABALHO == 'Rarely'), 'INTERFERE_NO_TRABALHO'] = 'Raramente'
data.loc[(data.INTERFERE_NO_TRABALHO == 'Never'), 'INTERFERE_NO_TRABALHO'] = 'Nunca'
data.loc[(data.INTERFERE_NO_TRABALHO == 'Sometimes'), 'INTERFERE_NO_TRABALHO'] = 'As Vezes'
```

```python
plt.title('Pessoas que tem doenças mentais e que isso os atrapalham na hora do trabalho')

data.groupby('INTERFERE_NO_TRABALHO').size().plot(kind='barh', figsize=(5,5),  color='#A020F0',rot=40)
```

```python
data.loc[(data.SUPERVISAO_MEDICA == 'Yes'), 'SUPERVISAO_MEDICA'] = 'Tem acompanhamento médico'
data.loc[(data.SUPERVISAO_MEDICA == 'No'), 'SUPERVISAO_MEDICA'] = 'Não tem acompanhamento médico'
data.loc[(data.SUPERVISAO_MEDICA == 'Some of them'), 'SUPERVISAO_MEDICA'] = 'Algumas vezes tem acompanhamento médico'
```

```python
plt.title('Pessoas com acompanhamento médico')

data.groupby('SUPERVISAO_MEDICA').size().plot(kind='barh', figsize=(5,5),  color='#A020F0',rot=40)
```

```python
data2 = data.copy()
data.loc[(data.GENERO == 'Male'), 'GENERO'] = 'Homens'
data.loc[(data.GENERO == 'male'), 'GENERO'] = 'Homens'
data.loc[(data.GENERO == 'M'), 'GENERO'] = 'Homens'
data.loc[(data.GENERO == 'Female'), 'GENERO'] = 'Mulheres'
data.loc[(data.GENERO == 'female'), 'GENERO'] = 'Mulheres'
data.loc[(data.GENERO == 'F'), 'GENERO'] = 'Mulheres'
data2.loc[(data.GENERO == 'Mulheres'), 'GENERO'] = 'Hetero'
data2.loc[(data.GENERO == 'Homens'), 'GENERO'] = 'Hetero'
data.loc[(data2.GENERO != 'Hetero'), 'GENERO'] = 'Outros' 
```

```python
plt.title('Gênero das pessoas com doenças mentais')

data.groupby('GENERO').size().plot(kind='bar', figsize=(5,5),  color='#A020F0',rot=40)
```
